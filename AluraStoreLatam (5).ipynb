##Distribución de categorías de productos

#### Datos en formato porcentual por tienda:

```python
def calcular_distribucion(tienda_df, nombre_tienda):
    """
    Para calcular distribución porcentual de categorías en una tienda
    """
    distribucion = (tienda_df['Categoría del Producto']
                    .value_counts(normalize=True)
                    * 100)

    return pd.DataFrame({
        'Tienda': nombre_tienda,
        'Categoría': distribucion.index,
        'Porcentaje': distribucion.values
    })

dist_tienda1 = calcular_distribucion(tienda, 'Tienda 1')
dist_tienda2 = calcular_distribucion(tienda2, 'Tienda 2')
dist_tienda3 = calcular_distribucion(tienda3, 'Tienda 3')
dist_tienda4 = calcular_distribucion(tienda4, 'Tienda 4')

distribucion_completa = pd.concat([dist_tienda1, dist_tienda2, dist_tienda3, dist_tienda4])
```

####Pivotear los datos para el gráfico

```python
pivot_data = distribucion_completa.pivot(
    index='Tienda',
    columns='Categoría',
    values='Porcentaje'
).fillna(0)
```

####Crear el gráfico de barras apiladas

```python
fig, ax = plt.subplots(figsize=(12, 8))

categorias = pivot_data.columns
colores = plt.cm.Set3(np.linspace(0, 1, len(categorias)))
bottom_vals = np.zeros(len(pivot_data))

for i, (categoria, color) in enumerate(zip(categorias, colores)):
    valores = pivot_data[categoria]
    ax.bar(pivot_data.index, valores, bottom=bottom_vals,
           label=categoria, color=color, edgecolor='black')
    bottom_vals += valores

ax.set_title('Distribución de Categorías de Productos por Tienda',
             fontsize=16, fontweight='bold')
ax.set_ylabel('Porcentaje (%)', fontsize=12)
ax.set_xlabel('Tienda', fontsize=12)
ax.set_ylim(0, 100)

ax.legend(title='Categorías', bbox_to_anchor=(1.05, 1), loc='upper left')

for bar in ax.containers:
    ax.bar_label(bar, label_type='center', fmt='%.1f%%',
                 fontsize=9, color='black')

plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

##1. Introducción
El presente informe tiene como objetivo analizar el desempeño de cuatro tiendas (Tienda 1, Tienda 2, Tienda 3 y Tienda 4) para recomendar cuál representa la mejor oportunidad de venta para el Sr. Juan. Para ello, se evaluaron los siguientes factores clave:

- Ingresos totales generados.

- Distribución de ventas por categoría de producto.

- Calificaciones promedio de los clientes.

- Productos más y menos vendidos.

- Costos de envío promedio.

El análisis se realizó utilizando datos históricos de ventas, y se respalda con visualizaciones que permiten una comparación objetiva entre las tiendas. La decisión final se basará en un equilibrio entre rentabilidad, satisfacción del cliente y eficiencia operativa.
